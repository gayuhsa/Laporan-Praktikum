```python
# Kelompok 6

daftar_mahasiswa = []
dictionary_mahasiswa = {}


class Mahasiswa:
  def __init__(self, nama, nim, nilai_akhir):
    if nama in dictionary_mahasiswa:
      print(f"Mahasiswa dengan nama {nama} sudah ada")
      return
    self.nama = nama
    self.nim = nim
    self.nilai_akhir = nilai_akhir
    daftar_mahasiswa.append(self)
    dictionary_mahasiswa[nama] = self

  def tambah_nilai(self, nilai_baru):
    self.nilai_akhir = nilai_baru

  def info_mahasiswa(self):
    return f"Nama\t\t: {self.nama}\nNIM\t\t: {self.nim}\nNilai akhir\t: {self.nilai_akhir}\n"


def tampilkan_semua_mahasiswa():
  for v in dictionary_mahasiswa:
    mahasiswa = dictionary_mahasiswa[v]
    print(mahasiswa.info_mahasiswa())

def cari_mahasiswa(nim):
  for data in daftar_mahasiswa:
    if data.nim == nim:
      print(f"Nama\t\t: {data.nama}\nNIM\t\t: {data.nim}\nNilai akhir\t: {data.nilai_akhir}")


print("""Daftar perintah:
1. Tambah mahasiswa
2. Update nilai
3. Tampilkan semua data mahasiswa
4. Cari mahasiswa berdasarkan NIM
5. Hentikan program
""")


while True:
  try:
    command = input()

    if command == "5":
      break

    if command == "1":
      nama = input("Nama\t\t: ")
      nim = int(input("NIM\t\t: "))
      nilai_akhir = int(input("Nilai akhir\t: "))
      mahasiswa_baru = Mahasiswa(nama, nim, nilai_akhir)
    elif command == "2":
      nama = input("Nama\t\t: ")

      if nama in dictionary_mahasiswa:
        nilai_baru = int(input("Nilai\t\t: "))
        dictionary_mahasiswa[nama].tambah_nilai(nilai_baru)
        print(f"Berhasil mengganti nilai {nama} menjadi {nilai_baru}")
      else:
        print(f"Mahasiswa dengan nama {nama} tidak ditemukan")
    elif command == "3":
      tampilkan_semua_mahasiswa()
    elif command == "4":
      cari_mahasiswa(int(input("NIM\t\t: ")))
    else:
      print("Perintah harus angka 1 - 5")

    print("\n")

  except ValueError:
    print("Error\n")
```
